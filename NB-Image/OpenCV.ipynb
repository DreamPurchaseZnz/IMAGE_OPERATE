## 图像直方图处理
### 灰度图像处理

```python
import cv2
import numpy as np
import matplotlib.pylab as plt
```

```python
img = cv2.imread('Desert.jpg',0)
print(img.shape)
```

```python
gray = np.array(img).flatten()
```

```python
n,bins,patches = plt.hist(gray ,bins=256,normed = 1,\
                          facecolor = 'blue',alpha = 0.75)
plt.show()
```

###  彩色图像处理

```python
img = cv2.imread('Desert.jpg',3)
print(img.shape)
```

```python
r,g,b = cv2.split(img)
print(r.shape,g.shape)
```

```python
cv2.imshow('r',r)
cv2.waitKey(0)
```

```python
ar = np.array(r).flatten()
ag = np.array(g).flatten()
ab = np.array(b).flatten()
```

```python
n,bins,patches = plt.hist(ar,bins=256, \
                          normed=1,facecolor='r',edgecolor = 'r',hold=1)
```

```python
plt.hist(ag,bins=256,normed=1,\
                         facecolor='g',edgecolor = 'g',hold=1);
```

```python
plt.hist(ab,bins=256,normed=1,\
                         facecolor='b',edgecolor = 'b');
```

```python
plt.show()
```

## Canny边缘检测

```python
r,b,g= cv2.split(img)
```

```python
edge = cv2.Canny(g,30,60)
print('Size:',edge.shape)
```

```python
cv2.imshow('edge',edge)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.imshow(edge,cmap ='gray')
plt.axis('off')
plt.show()
```
