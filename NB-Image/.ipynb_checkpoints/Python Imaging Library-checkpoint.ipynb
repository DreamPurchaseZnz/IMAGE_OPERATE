## Image Enhance

```python
from PIL import ImageEnhance
```

```python
enhancer = ImageEnhance.Sharpness(img)
```

```python
fig = plt.figure()
for i in range(8):
    factor = i/4.0
    img_enhance = enhancer.enhance(factor)
    img_new = np.array(img_enhance)
    ax = fig.add_subplot(4,2,i+1)
    ax.imshow(img_new)
    ax.axis('off')
    
plt.show()
```

```python
from PIL import Image

from pylab import *

# 读取图像到数组中，并灰度化
im = array(Image.open('45.jpg').convert('L'))

#显示时抛弃颜色信息
gray()

# 显示轮廓图像
contour(im, origin='image')

# 在原点的左上角显示
axis('equal')

#关闭坐标轴
axis('off')

show()
```

## 使用Matplotlib绘制直方图
>图像的直方图用来表征该图像像素值的分布情况。用一定数目的小区间(bin)来指定表征像素值的范围,每个小区间会得到落入该小区间表示范围的像素数目。(灰度)图像的直方图可以使用 hist() 函数绘制

>hist() 只接受**一维数组**作为输入,所以我们在绘制图像直方图之前,必须先对图像进行**压平处理**。flatten() 方法将任意数组按照行优先准则转换成一维数组

```python
from PIL import Image
from matplotlib.pylab import *

# 读取图像到数组中，并灰度化
im = array(Image.open('Desert.jpg').convert('L'))

# 直方图图像
hist(im.flatten(),128)

# 显示
show()
```

```python
from PIL import Image,ImageDraw,ImageFilter
import matplotlib.pylab as plt
import numpy as np
img = Image.open("Desert.jpg")
img.show()
```

```python
img = img.convert("RGB") 
```

* 经过PIL自带filter处理

```python
imgfilted_1 = img.filter(ImageFilter.BLUR)
imgfilted_2 = img.filter(ImageFilter.CONTOUR)
imgfilted_3 = img.filter(ImageFilter.EDGE_ENHANCE)
imgfilted_4 = img.filter(ImageFilter.EDGE_ENHANCE_MORE)
imgfilted_5 = img.filter(ImageFilter.EMBOSS)                    
imgfilted_6 = img.filter(ImageFilter.FIND_EDGES)                                                
imgfilted_7 = img.filter(ImageFilter.SMOOTH)
imgfilted_8 = img.filter(ImageFilter.SMOOTH_MORE)
imgfilted_9 = img.filter(ImageFilter.SHARPEN)
imgfilted_10 = img.filter(ImageFilter.DETAIL)
```

* 组合使用filter

```python
group_imgfilted = img.filter(ImageFilter.CONTOUR)
group_imgfilted = group_imgfilted.filter(ImageFilter.SMOOTH_MORE)
```

* 图像显示
> locals()[varName]

> vars()[varName]

> eval(varName)实现字符串匹配相应变量


```python
fig = plt.figure()
for i in range(10):
    string = 'imgfilted_%d'%(i+1)
    img_new = np.array(locals()[string])
    ax = fig.add_subplot(5,2,i+1)
    ax.imshow(img_new)
    ax.axis('off')
    
plt.show()
```
